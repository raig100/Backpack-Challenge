```python
#drop na for train dataset
df_normalized[numerical] = df[numerical]
df_normalized = df_normalized.dropna()
df_normalized.isna().sum()
```

```python
# initiating x and y for train cols
x = df_normalized.drop(columns = ['Price'], axis = 1)
y = df_normalized['Price']
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train.head()
```

```python
#initiating model
model = ElasticNet(random_state = 42)
model
```

```python
#params of the model
model.get_params()
```

```python
#train model resuls
model.fit(x_train, y_train)
model.score(x_train, y_train)
```

```python
#train model result
pred = model.predict(x_test)
print(mean_absolute_error(y_test, pred))
```

```python
#train model results
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
rmse
```
